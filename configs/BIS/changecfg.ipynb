```python
import os, os.path as osp
ori_dataset_name = 'XQpink'
dataset_names = ['XQpurple', 'XQblue', 'XQXY', 'XQY']
cfg_path = f'./{ori_dataset_name}/'
files = os.listdir(cfg_path)
for file in files:
    if file.endswith('.py'):
        ori_cfg_path = osp.join(cfg_path , file)
        for dataset_name in dataset_names:
            new_cfg_path = ori_cfg_path.replace(ori_dataset_name, dataset_name)
            if not os.path.exists(path=os.path.dirname(new_cfg_path)):
                os.makedirs(os.path.dirname(new_cfg_path))
            with open(ori_cfg_path, "r") as f:
                srcCfg = f.read().split("\n")
                for i in range(len(srcCfg)):
                    srcCfg[i] = srcCfg[i].replace(ori_dataset_name, dataset_name)
                f.close()
                with open(new_cfg_path, "w") as f:
                    f.write('\n'.join(srcCfg))
                    # input(new_cfg_path)
```
