```python
import os
import time 

# y = time.ctie(os.stat("D:\\test\\1.txt").st_mtime)
# y
filename = '/home/yebh/mmdet2/eight_test.sh'
# read the last time to adjust the file

import os
import time

file_path = filename 

# 获取文件状态信息
file_stat = os.stat(file_path)

# 获取最后修改时间
# last_modified_time = time.ctime(file_stat.st_mtime)
last_modified_time = time.localtime(file_stat.st_mtime)

now_time = time.localtime(time.time())
print(now_time)
# compute the time difference
time_difference = (now_time.tm_hour - last_modified_time.tm_hour) * 60 + (now_time.tm_min - last_modified_time.tm_min)
print(time_difference)
# print("最后修改时间：", last_modified_time)
```
